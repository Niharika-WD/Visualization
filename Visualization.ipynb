# **DATA ANALYSIS AND VISUALIZATION PROJECT**

## *BOSTON CRIME DATASET*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
import pylab
```

```python
dataset=pd.read_csv("data/crime.csv",
                   encoding="latin1")
```

```python
dataset=dataset.replace(np.NaN,0)
dataset=dataset.replace("nan","no")
dataset=dataset.fillna(0)
```

## ***YEAR WISE***

```python
sns.set(style="whitegrid")
bax=sns.barplot(x=dataset["YEAR"].value_counts().index,y=dataset["YEAR"].value_counts().values,palette="Pastel2")
```

### The above barplot depicts the crime rate over the years.The crime rate spiked in 2015 from ~55,000 to ~90,000 in 2016.
### There was a decline in the crime rate in 2018

## ***MONTH WISE***

```python
sns.set(style="whitegrid")
bax=sns.barplot(x=dataset["MONTH"].value_counts().index,y=dataset["MONTH"].value_counts().values,palette="winter_r")
```

### The above barplot depicts the crime distribution over the months.The crime rate was highest in months of July and August ~33,000.

## ***WEEK DAY***

```python
sns.set(style="whitegrid")
bax=sns.barplot(x=dataset["DAY_OF_WEEK"].value_counts().index,y=dataset["DAY_OF_WEEK"].value_counts().values,palette="copper")
```

### The graph shows which days of the week show the highest number of crime. Unsuprisingly,its FRIDAY with ~48000

## ***HOUR WISE***

```python
sns.set(style="whitegrid")
bax=sns.barplot(x=dataset["HOUR"].value_counts().index,y=dataset["HOUR"].value_counts().values,palette="ocean_r")
```

### The graph shows that between 4p.m. and 6p.m. the crime is at its highest and its safer in the morning hours(3a.m. to 5a.m.)

## **UNIFORM CRIME REPORT**

```python
labels=dataset["UCR_PART"].value_counts().index
values=dataset["UCR_PART"].value_counts().values
color=["pink","purple","yellow","blue"]
plt.figure(figsize=(4,4))
plt.pie(values,
        labels=labels,
        autopct="%0.1f%%",
        colors=color,
        shadow=True,
        radius=1.2,
        labeldistance=1.2,
        pctdistance=0.5,
       textprops={"color":"black"})
plt.axis("equal")       
```

### The above pie-chart shows the percentage of different categories of UNIFORM CRIME REPORT(described as UCR_PART),namely- Part One ,Part Two, Part Three and others.
